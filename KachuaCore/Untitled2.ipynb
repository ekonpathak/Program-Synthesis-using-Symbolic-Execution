## Question1: Spectrogram of Audio File (25 points)

### Step-1: Let's import all the required libraries

```python
import os
import matplotlib.pyplot as plt

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd

audio_fpath = "/Users/lucifer/Desktop/untitled folder 3/"
audio_clips = os.listdir(audio_fpath)
print("No. of .wav files in audio folder = ",len(audio_clips))
```

### Step-2: Load audio file and visualize its waveform (using librosa)

```python
x, sr = librosa.load(audio_fpath+audio_clips[0], sr=44100)

print(type(x), type(sr))
print(x.shape, sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)
```

### Step-3: Convert the audio waveform to spectrogram

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

### Step-4: Applying log transformation on the loaded audio signals

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```
